```python
#get images and meta data
import os
from mistralai import Mistral
from PIL import Image

client = Mistral(api_key=api_key)

ocr_response = client.ocr.process(
    model="mistral-ocr-latest",
    document={
        "type": "document_url",
        "document_url": "https://arxiv.org/pdf/2406.15758"
    },
    include_image_base64=True
)


import os
import re
import json
import base64
import io



def extract_figure_tag(markdown, image_id):
    """Extracts just 'Figure X' (not full caption) following the image reference."""
    pattern = rf"!\[.?\]\({re.escape(image_id)}\)\s[\r\n]+(Figure\s+\d+)"
    match = re.search(pattern, markdown, re.IGNORECASE)
    return match.group(1).strip() if match else None

def save_figures_from_ocr(ocr_response, output_folder="data/figures"):
    output_folder = os.path.abspath(output_folder)
    os.makedirs(output_folder, exist_ok=True)
    figures_metadata = {}

    for page in ocr_response.pages:
        page_index = getattr(page, "index", None)
        markdown = getattr(page, "markdown", "")
        images = getattr(page, "images", [])

        for i, image in enumerate(images):
            image_id = getattr(image, "id", f"figure_{page_index}_{i}.jpeg")
            img_data = getattr(image, "image_base64", None)

            # Remove header if present
            if ',' in img_data:
                img_data = img_data.split(',', 1)[1]

            # Decode base64 to image
            img_bytes = base64.b64decode(img_data)
            img = Image.open(io.BytesIO(img_bytes))

            # Ensure image filename doesn't include subfolders
            image_filename = os.path.basename(image_id)
            save_path = os.path.join(output_folder, image_filename)

            if img_data:
                img.save(save_path)

                figure_key = extract_figure_tag(markdown, image_id) or f"Figure_{page_index}_{i}"

                figures_metadata[figure_key] = save_path.replace("\\", "/")
                  
    metadata_path = os.path.join(output_folder, "figures_metadata.json")
    with open(metadata_path, "w") as meta_file:
        json.dump(figures_metadata, meta_file, indent=4)

    print(f"✅ Saved {len(figures_metadata)} figures to '{output_folder}'")
    # print(f"📄 Metadata JSON saved to '{metadata_path.replace('\\', '/')}'")
    return figures_metadata


save_figures_from_ocr(ocr_response)
```

```python
from pptx import Presentation
from pptx.enum.shapes import PP_PLACEHOLDER, MSO_SHAPE_TYPE

def extract_pptx_layout_to_md(pptx_path: str, output_md: str = "pptx_layout_details.md"):
    """
    Extracts slide layout details from a PowerPoint template and writes them to a Markdown file.

    Args:
        pptx_path (str): Path to the input .pptx template file.
        output_md (str): Path to save the output .md file.
    """
    presentation = Presentation(pptx_path)

    with open(output_md, "w", encoding="utf-8") as f:
        for layout_idx, slide_layout in enumerate(presentation.slide_layouts):
            layout_name = slide_layout.name
            master_name = slide_layout.slide_master.name
            
            f.write(f"\n---\n\n## 🧩 Slide Layout {layout_idx}: {layout_name}\n")
            f.write(f"**🎨 Master Theme:** `{master_name}`\n\n")

            for shape in slide_layout.shapes:
                shape_name = shape.name
                shape_type = shape.shape_type
                shape_type_name = shape_type.name if isinstance(shape_type, MSO_SHAPE_TYPE) else str(shape_type)
                
                width = shape.width
                height = shape.height
                left = shape.left
                top = shape.top

                if shape.is_placeholder:
                    phf = shape.placeholder_format
                    try:
                        ph_type = PP_PLACEHOLDER(phf.type).name
                    except ValueError:
                        ph_type = str(phf.type)

                    f.write(f"### 🔲 Placeholder: `{shape_name}`\n")
                    f.write(f"- **Placeholder idx:** `{phf.idx}`\n")
                    f.write(f"- **Type:** `{ph_type}`\n")
                    f.write(f"- **Shape Type:** `{shape_type_name}`\n")
                    f.write(f"- **Size:** `{width} x {height}`\n")
                    f.write(f"- **Position:** `{left} x {top}`\n\n")

                else:
                    f.write(f"### 🖼️ Shape Layer: `{shape_name}` (Not a placeholder)\n")
                    f.write(f"- **Shape Type:** `{shape_type_name}`\n")
                    f.write(f"- **Size:** `{width} x {height}`\n")
                    f.write(f"- **Position:** `{left} x {top}`\n\n")

    print(f"✅ Markdown with full layout details saved to: {output_md}")
```
